# This notebook will walk through creating an animated chloropleth map of PA using COVID data from Johns Hopkins

```python
# Import libraries
import numpy as np 
import pandas as pd 
import plotly as py
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
# Read in US data
df = pd.read_csv("D:/GitHub/AI/DataScience/Visualization/COVID/data/time_series_covid19_confirmed_US.csv")
```

```python
df.head()
```

```python
df_pa = df[(df.Province_State == "Pennsylvania")]
```

```python
df_pa.head()
```

```python
#Creating the visualization
fig = px.choropleth(df, 
                    locations="Province_State", 
                    locationmode = "State names",
                    color="Confirmed", 
                    hover_name="Province_State", 
                    animation_frame="Date"
                   )
fig.update_layout(
    title_text = 'PA Spread of Coronavirus',
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
    ))
    
fig.show()
```
